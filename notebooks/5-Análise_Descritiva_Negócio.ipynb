## Tratamento de dados

```python
colunas = df_vendas.columns
colunas_gerente = df_gerentes.columns
colunas_dez = df_vendas_dez.columns
print(colunas)
print(colunas_dez)
print(colunas_gerente)
# Padronizando as colunas
colunas = list(i.strip().replace(' ', '_').lower() for i in colunas)
colunas_dez = list(i.strip().replace(' ', '_').lower() for i in colunas_dez)
colunas_gerente = list(i.strip().replace(' ', '_').lower() for i in colunas_gerente)
```

```python
#Aplicando
df_vendas.columns = colunas
df_vendas_dez.columns = colunas_dez
df_gerentes.columns = colunas_gerente
print(colunas)
print(colunas_dez)
print(colunas_gerente)
```

```python
#Apagando a coluna índice
df_vendas = df_vendas.drop(columns=['indice'])
```

```python
#Removendo linhas duplicadas
df_vendas.drop_duplicates(inplace=True)
df_vendas_dez.drop_duplicates(inplace=True)
```

```python
df_vendas['valor_unitário'] = df_vendas['valor_unitário'].str.replace(r'R\$', '', regex=True)# Removido o R$
df_vendas['valor_unitário'] = df_vendas['valor_unitário'].astype(float)
```

```python
df_vendas.fillna({'valor_final': df_vendas['quantidade']*df_vendas['valor_unitário']}, inplace=True) #Cálculo coluna valor_final
df_vendas.fillna({'custo_do_produto': df_vendas['valor_final']*.6}, inplace=True) #Cálculo coluna custo_do_produto
display(df_vendas)
```

```python
df_vendas.dropna(inplace=True)
df_vendas.isna().sum()
```

```python
df_faturamento_total_base = df_vendas['valor_final'].sum()
print(f'Faturamento df_vendas: R${df_faturamento_total_base:,.2f}')
```

```python
df_faturamento_total_dez = df_vendas_dez['valor_final'].sum()
print(f'Faturamento df_vendas_dez: R${df_faturamento_total_dez:,.2f}')
```

## Concat

```python
df_vendas_final = pd.concat([df_vendas, df_vendas_dez], axis=0, join='inner')
```

## Merge

```python
df_vendas_final = df_vendas_final.merge(df_gerentes, on='id_loja', how='outer')
```

### Qual foi o faturamento total no período analisado?

```python
df_faturamento_total = df_vendas_final['valor_final'].sum()
print(f'Faturamento total: R${df_faturamento_total:,.2f}')
```

### Qual loja teve o maior faturamento?

```python
maior_fat_loja = df_vendas_final.groupby('id_loja')['valor_final'].sum().reset_index().sort_values(by='valor_final', ascending=False)


fig = px.bar(maior_fat_loja, x='id_loja', y='valor_final', text=maior_fat_loja['valor_final'], title='Receita por Loja', color='id_loja')

fig.update_traces(textposition='outside', texttemplate='R$ %{text:0.3s}')

fig.show()
```

### Qual produto foi o mais vendido em quantidade?

```python
maior_prod_vend = df_vendas_final.groupby('produto')['quantidade'].sum().reset_index().sort_values(by='quantidade', ascending=False)


fig = px.bar(maior_prod_vend, x='produto', y='quantidade', title='Quantidade de produtos vendidos', color_discrete_sequence=px.colors.qualitative.Bold)

fig.update_traces(textposition='outside', texttemplate='R$ %{text:0.3s}')

fig.show()
```

### Qual produto foi o menos vendido em quantidade?

```python
menor_prod_vend = df_vendas_final.groupby('produto')['quantidade'].sum().reset_index().sort_values(by='quantidade', ascending=True)


fig = px.bar(menor_prod_vend, x='produto', y='quantidade', title='Quantidade de produtos vendidos', color_discrete_sequence=px.colors.qualitative.Bold)

fig.update_traces(textposition='outside', texttemplate='R$ %{text:0.3s}')

fig.show()
```

### Qual produto gerou a maior receita (somatório do Valor Final)?

```python
maior_prod_receita = df_vendas_final.groupby('produto')['valor_final'].sum().reset_index().sort_values(by='valor_final', ascending=False)

fig = px.bar(maior_prod_receita, x='produto', y='valor_final', text='valor_final', title='Produtos e suas receitas', color_discrete_sequence=px.colors.qualitative.Bold)

fig.update_traces(textposition='outside', texttemplate='R$ %{text:0.3s}')

fig.show()
```

### Qual produto gerou maior lucro?

```python
df_vendas_final['lucro'] = df_vendas_final['valor_final'] - df_vendas_final['custo_do_produto']
maior_prod_lucro = df_vendas_final.groupby('produto')['lucro'].sum().reset_index().sort_values(by='lucro', ascending=False)


fig = px.bar(maior_prod_lucro, x='produto', y='lucro', text='lucro', title='Produtos e seus lucros', color_discrete_sequence=px.colors.qualitative.Bold)

fig.update_traces(textposition='outside', texttemplate='R$ %{text:0.3s}')

fig.show()
```

### Qual a média da margem de lucro por produto?

```python
df_vendas_final['margem_lucro'] = (df_vendas_final['valor_final'] - df_vendas_final['custo_do_produto']) / df_vendas_final['valor_final']

media_lucro_prod = df_vendas_final.groupby('produto')['margem_lucro'].mean().reset_index().sort_values(by='margem_lucro', ascending=False)


fig = px.bar(media_lucro_prod, x='produto', y='margem_lucro', text='margem_lucro', title='Produtos e suas margens de lucro', color_discrete_sequence=px.colors.qualitative.Bold)

fig.update_traces(textposition='outside', texttemplate='%{text:0.1}')

fig.show()
```

### Qual gerente gerou maior receita?

### Qual gerente gerou o maior lucro?

```python
df_lucro_gerente = df_vendas_final.groupby('gerente')['lucro'].sum().reset_index().sort_values(by='lucro',ascending=False)

display(df_lucro_gerente)
```

### Qual foi o mês com maior faturamento?

```python
df_vendas_final['ano - mes'] = df_vendas_final['data'].dt.to_period('M')
```

```python
df_fat_mes = df_vendas_final.groupby('ano - mes')['valor_final'].sum().reset_index()

df_fat_mes['ano - mes'] = df_fat_mes['ano - mes'].dt.to_timestamp()
fig = px.line(df_fat_mes, x='ano - mes', y='valor_final')

fig.show()
```

```python
df_vendas_final['mes'] = df_vendas_final['data'].dt.month
```

```python
df_gerentes_mes = df_vendas_final.groupby(['gerente', 'mes'])['valor_final'].sum().reset_index()

fig = px.line(
    df_gerentes_mes,
    x='mes',
    y='valor_final',
    color='gerente',
    title='Vendas Mensais por Gerente',
    height=500,
    width=1000,
    markers=True
)

fig.update_yaxes(title='Vendas', ticks='outside', tickprefix='R$ ')

fig.show()
```
